```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = "Times New Roman"
```

```python
x = np.linspace(0,2*np.pi, 100)
def function(a,b):
    y=a*np.exp(-b*x)
    return(y)
```

```python
plt.plot(x,function(2,0.5), c = 'black')
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $')
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $')
plt.savefig("Graph_of_an_Exponential_Function.pdf", format="pdf")
```

```python
m = (-0.9/(np.pi))
b = 2
sigma = m*x+b
scatter_func=np.random.normal(function(2,0.5), sigma)
plt.scatter(x,scatter_func, c = 'teal')
plt.plot(x,function(2,0.5), c = 'black')
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $ (With Scatter)')
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $ (With Scatter)')
```

```python
plt.errorbar(x,function(2,0.5), sigma, 0,marker = ".", ls='-', c = "black", ecolor = "teal")
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $ (With Error)')
plt.savefig('Graph_of_an_Exponential_Function_with_Error.pdf',format="pdf")
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $ (With Error)')
```

```python
#PART 3
from scipy.optimize import curve_fit
```

```python
#for y=a*np.exp(-b*x)
x = np.linspace(0,5,100)
def y_scatter(x, a,b):
    y = a*np.exp(-b*x)
    return(y)
y_scatter_error= np.random.normal(y_scatter(x,1,1),0.2)
plt.scatter(x,y_scatter_error)
y_scatter_error_estimate = curve_fit(y_scatter, x, y_scatter_error)
print(y_scatter_error_estimate)
plt.plot(x,y_scatter(x,1.12741474,1.05355981), color= 'red')
```

```python
#for y=a/(x**b)
x1 = np.linspace(0.1,2,50)
def y1_scatter(x1,a,b):
    y1 = a/(x1**b)
    return(y1)
y1_scatter_error= np.random.normal(y1_scatter(x1,1,1),0.5)
plt.scatter(x1,y1_scatter_error)
y1_scatter_error_estimate = curve_fit(y1_scatter, x1, y1_scatter_error)
print(y1_scatter_error_estimate)
plt.plot(x1, y1_scatter(x1,1.00084668,1.01224825),c="red")
```

```python
'''
PART 3: Modeling a scatterplot using curve_fit
    - Create a scatter plot with error, making a shape
    - fit it with an exponential and a power model
    - Print the optimized values of a and b using an f-string
'''
#creating the scatter plot
x2 = np.linspace(0.5, 10, 60) # 60 evenly spaced points from 0.5 to 10 (x)
y2 = (1*np.exp(-1*x2))
y2_error = np.random.normal((y2),0.1) #making the model exponential, with error
plt.scatter(x2,y2_error) #plotting it
```

```python
def model_curve_exp(x2,a,b): #first model curve using an exponential func
    y3 = a*np.exp(b*x2)
    return(y3)
def model_curve_pow(x2,a,b): #second model curve using a power func
    y3 = a/(x2**b)
    return(y3)
plt.scatter(x2,y2_error) #plotting it
mcexp_fit_params, mcexp_fit_cov = curve_fit(model_curve_exp,x2,y2_error)
mcpow_fit_params, mcpow_fit_cov = curve_fit(model_curve_pow,x2,y2_error)
print(f'The a and b values of the exponential model are {mcexp_fit_params[0]:.3f} and {mcexp_fit_params[1]:.3f}')
print(f'The a and b values of the power model are {mcpow_fit_params[0]:.3f} and {mcpow_fit_params[1]:.3f}')
plt.plot(x2,model_curve_exp(x2,mcexp_fit_params[0],mcexp_fit_params[1]), c='red', label = r'$\ ae^{bx} $')
plt.plot(x2,model_curve_pow(x2,mcpow_fit_params[0],mcpow_fit_params[1]), c='orange', label = r'$\ ax^{-b} $')
plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.savefig("Modelling_an_Exponential_Function_with_Two_Functions.pdf", format="pdf")
```
