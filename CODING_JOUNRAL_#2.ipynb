```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = "Times New Roman"
```

```python
x = np.linspace(0,2*np.pi, 100)
def function(a,b):
    y=a*np.exp(-b*x)
    return(y)
```

```python
plt.plot(x,function(2,0.5), c = 'maroon')
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $')
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $')
plt.savefig("Graph_of_an_Exponential_Function.pdf", format="pdf")
```
