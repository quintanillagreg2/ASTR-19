```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = "Times New Roman"
```

```python
x = np.linspace(0,2*np.pi, 100)
def function(a,b):
    y=a*np.exp(-b*x)
    return(y)
```

```python
plt.plot(x,function(2,0.5), c = 'black')
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $')
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $')
plt.savefig("Graph_of_an_Exponential_Function.pdf", format="pdf")
```

```python
m = (-0.9/(np.pi))
b = 2
sigma = m*x+b
scatter_func=np.random.normal(function(2,0.5), sigma)
plt.scatter(x,scatter_func, c = 'teal')
plt.plot(x,function(2,0.5), c = 'black')
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $ (With Scatter)')
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $ (With Scatter)')
```

```python
plt.errorbar(x,function(2,0.5), sigma, 0,marker = ".", ls='-', c = "black", ecolor = "teal")
plt.xlabel(r'$x$')
plt.ylabel(r'$\ -2e^{-0.5x} $ (With Error)')
plt.savefig('Graph_of_an_Exponential_Function_with_Error.pdf',format="pdf")
plt.title(f'Graph of an Exponential Function from 0 to' r'$\ 2\pi $ (With Error)')
```
