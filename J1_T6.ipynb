```python
import numpy as np
```

```python
def sin(x): #defines sin function
    return np.sin(x) #makes the sin function work using numpy
```

```python
def cos(x): #defines cos function
    return np.cos(x) 
```

```python
x = np.linspace(0, 2*np.pi, 1000) #createes 1000 evenly spaces values between
```

```python
print("x    sin(x)    cos(x)")#prints the title, spaced out
for i in range(10): #prints the first 10 values
    x_val = int(x[i]*1000)/1000 #turns the first 4 digits into an integer and then divides by 1000 putting it into a float
    sin_val = int(sin(x[i])*1000)/1000
    cos_val = int(cos(x[i])*1000)/1000
    print(x_val, sin_val, cos_val) #puts it all togvether
```
